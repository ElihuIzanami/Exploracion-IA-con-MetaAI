Ese concepto se llama **broadcasting**. El broadcasting es un mecanismo que permite a NumPy realizar operaciones en arrays de diferentes dimensionalidades.

````python
arreglo
````

````python
# Realizar broadcasting con un array a través de producto y suma

arreglo + 1.6 # [1.6, 1.6, 1.6, 1.6, 1.6]
````

### **MATRICES**

### **PLOT**

Un **plot** en Matplotlib es la **herramienta fundamental** para visualizar datos numéricos de manera gráfica. Es decir, apoya a entender la relación entre dos variables de forma intuitiva.

#### **PLANO CARTESIANO**

Una gráfica de estilo **plano cartesiano**, también conocida como gráfica de coordenadas cartesianas o simplemente gráfica, es una representación visual de datos numéricos en un sistema de coordenadas formado por dos ejes perpendiculares: el eje de las abscisas (horizontal) y el eje de las ordenadas (vertical). Cada punto en la gráfica corresponde a un par de valores (x, y), y al unir estos puntos se pueden obtener líneas, curvas o puntos aislados que muestran la relación entre las variables representadas en los ejes.

````python
# Definir variables

x = np.linspace(-10, 10, 50)
y = np.square(x)
````

````python
x
````

````python
print(x)
````

````python
# Realizar parábola

plt.plot(x, y, color = '#29FF4B', label = 'Función cuadrada')
plt.axhline(y = 0, color = 'red')
plt.axvline(x = 0, color = 'red')
plt.title('Parábola')
plt.xlabel('Eje de las x')
plt.ylabel('Eje de las y')
plt.legend(loc = 'upper left') # lower - abajo
plt.grid()
plt.show()
````

#### **HISTOGRAMA**

Un **histograma** es una representación gráfica de la distribución de frecuencia de un conjunto de datos continuos. Se construye dividiendo el rango de valores de los datos en intervalos iguales (llamados clases o bins) y contando el número de observaciones que caen en cada intervalo. Cada intervalo se representa como una barra vertical, cuya altura es proporcional al número de observaciones en ese intervalo. Los histogramas son útiles para visualizar la forma de la distribución de los datos y detectar patrones como la simetría, la asimetría o la presencia de valores atípicos.

````python
# Definir variable

x = np.random.randn(1000)
````

````python
# Realizar histograma

plt.hist(x, bins = 20, color = 'pink', alpha = 0.8, label = 'Distribución Normal')
plt.axhline(y = 0, color = 'red')
plt.title('Histograma')
plt.xlabel('Datos')
plt.ylabel('Frecuencia')
plt.legend(loc = 'upper right')
plt.grid()
plt.show()
````

#### **GRÁFICO DE DISPERSIÓN**

Un **gráfico de dispersión** es una representación gráfica de la relación entre dos variables numéricas. Cada punto en el gráfico corresponde a un par de valores (x, y), y la posición del punto en el plano cartesiano indica los valores de ambas variables para esa observación. Los gráficos de dispersión son útiles para identificar patrones de asociación entre las variables, como la correlación lineal o no lineal, y detectar la presencia de valores atípicos.

````python
# Definir variables

np.random.seed(1)
x = np.random.rand(100)
y = 2 * x + 0.1 * np.random.randn(100)
````

````python
# Realizar gráfico de distribución

plt.scatter(x, y, color = 'gold', s = 20, label = 'Correlación lineal')
plt.axhline(y = 0, color = 'red')
plt.axvline(x = 0, color = 'red')
plt.title('Dispersión')
plt.xlabel('Eje de las x')
plt.ylabel('Eje de las y')
plt.legend(loc = 'upper left')
plt.grid()
plt.show()
````

#### **SUBPLOT**

Un **subplot** te permite organizar múltiples plot en una misma figura. Los subplot son muy útiles cuando tienes varios conjuntos de datos relacionados y quieres analizarlos de forma conjunta.

````python
# Definir variables

x = np.linspace(-10, 10, 50)
y = np.square(x)
````

````python
# Realizar gráfico dividido

plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
plt.subplot(1, 2, 2)
plt.hist(y)
plt.show()
````

````python
plt.subplot(2, 1, 1)
plt.plot(x, y, 'r')
plt.subplot(2, 1, 2)
plt.hist(y)
plt.show()
````
