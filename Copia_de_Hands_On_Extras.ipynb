```python
def lectura_comentarios(comentarios, nivel=0):
  for comentario in comentarios:
    sangria = "      " * nivel
    print(f"{sangria}----{comentario['usuario']} dice {comentario['texto']}")

lectura_comentarios(comentarios)
```

```python
def lectura_comentarios(comentarios, nivel=0):
  for comentario in comentarios:
    sangria = "      " * nivel
    print(f"{sangria}----{comentario['usuario']} dice {comentario['texto']}")
    for respuesta in comentario['respuestas']:
      sangria = "      " * nivel
      print(f"{sangria}----{respuesta['usuario']} dice {respuesta['texto']}")

lectura_comentarios(comentarios)
```

```python
def lectura_comentarios(comentarios, nivel=0):
  for comentario in comentarios:
    sangria = "      " * nivel
    print(f"{sangria}----{comentario['usuario']} dice {comentario['texto']}")

    if comentario['respuestas']:
      lectura_comentarios(comentario['respuestas'], nivel + 1)

lectura_comentarios(comentarios)
```
